```python
#Number of columns
columns =df.columns
print(columns)
```

```python
#Calculate the mean rating
mean_rating = df['IMDB_Rating'].mean()
print(mean_rating)
```

```python
#counting how many unique genre there are
count_uniq = df['Genre'].nunique()
print(count_uniq)
```

```python
#Count of each unique genre
unique_generes = df['Genre'].value_counts()
unique_generes.to_csv('uniq_genre.csv')
print(unique_generes)
```

# Data Visualization


* Plot a histogram of the genre distribution of the movies. How many movies from each genre are there in this list.
* Plot a scatterplot in this dataset for the number of movies vs year (Each year, how many movies made to this top 1000 list?)




```python
import matplotlib.pyplot as plt

# Plot the histogram
plt.bar(unique_generes.index, unique_generes.values)

# Customize the plot
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.title('Genre Distribution of Movies')
plt.xticks(rotation=45)
plt.subplots_adjust(bottom=0.5)
plt.xticks(fontsize=5)

# Display the plot
plt.show()
```

```python
# Group the movies by year and count the number of movies in each year
movies_by_year = df.groupby('Released_Year').size().reset_index(name='num_movies')

# Plot the scatterplot
plt.scatter(movies_by_year['Released_Year'], movies_by_year['num_movies'])

# Customize the plot
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.title('Number of Movies per Year')
plt.subplots_adjust(bottom=0.25)
plt.xticks(fontsize=5)
plt.xticks(rotation=45)

# Display the plot
plt.show()
```
